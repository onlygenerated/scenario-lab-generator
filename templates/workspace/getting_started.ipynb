## Database Connections

| Database | Host | Port | Database | User | Password |

```python
import pandas as pd
from sqlalchemy import create_engine

# Connection strings
source_engine = create_engine('postgresql://labuser:labpass@source-db:5432/source_db')
target_engine = create_engine('postgresql://labuser:labpass@target-db:5432/target_db')

print('Engines created successfully!')
```

```python
# Quick test: list tables in source database
source_tables = pd.read_sql_query(
    "SELECT table_name FROM information_schema.tables WHERE table_schema = 'public'",
    source_engine
)
print('Source tables:')
source_tables
```

```python
# Quick test: list tables in target database
target_tables = pd.read_sql_query(
    "SELECT table_name FROM information_schema.tables WHERE table_schema = 'public'",
    target_engine
)
print('Target tables:')
target_tables
```
